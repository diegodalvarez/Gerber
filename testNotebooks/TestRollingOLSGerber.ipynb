# OLS Varaible estimation

In this case we'll be using the standard calculation to the variables, except we are going to use Gerber Covariance in lieu of Covariance, the calculation is analog to simple linear regression

```python
exog = df_tmp.ADP #y
endog = df_tmp.ADBE #x
window = 30
```

```python
df_params = Gerber().RollingGerberOLS(endog = endog, exog = exog, window = 30, verbose = True)
```

```python
df_params.set_index("date").plot(subplots = True, figsize = (20,6))
plt.tight_layout()
plt.show()
```
